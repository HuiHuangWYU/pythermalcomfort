# Demo: Threshold

This notebook demonstrates the new `Threshold` class for fixed-assumption threshold-region plots.

```python
import matplotlib.pyplot as plt

from pythermalcomfort.models import pmv_ppd_iso, utci
from pythermalcomfort.plots.threshold import Threshold, ThresholdsConfig
```

## Build A Threshold Object

The builder-style workflow stays simple: define the x-axis, the y-axis, and the fixed assumptions.

```python
Model_1 = (
    Threshold(model_func=pmv_ppd_iso)
    .x(tdb=(18.0, 34.0))
    .y(rh=(20.0, 100.0))
    .parameters(vr=0.10, met=1.2, clo=0.5, wme=0.0)
    .plot(
        output="pmv",
        thresholds=ThresholdsConfig(thresholds=[-0.5, 0.5]),
        x_resolution=0.2,
        y_resolution=0.5,
    )
)
```

## Customizable Threshold Plot

This example uses the customizable region labels and colors.

```python
Model_1 = (
    Threshold(model_func=pmv_ppd_iso)
    .x(tdb=(18.0, 34.0))
    .y(rh=(20.0, 100.0))
    .parameters(vr=0.10, met=1.2, clo=0.5, wme=0.0)
    .plot(
        output="pmv",
        thresholds=ThresholdsConfig(
            thresholds=[-0.5, 0.5],
            labels=["Cold", "Neutral", "Hot"],
            colors=["#4c78a8", "#d9d9d9", "#c9415f"],
        ),
        x_resolution=0.2,
        y_resolution=0.5,
        title="PMV Threshold Regions",
        line_kws={"color": "black", "linewidth": 2.0, "linestyle": "dotted"},
        fill_kws={"alpha": 0.65},
    )
)
```

## Legend And Boundary-Line Options

The left panel turns the legend off. The right panel keeps the legend but hides the boundary lines.

```python
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(14, 5), constrained_layout=True)

Threshold(model_func=pmv_ppd_iso).x(tdb=(18.0, 34.0)).y(rh=(20.0, 100.0)).parameters(
    vr=0.10, met=1.2, clo=0.5, wme=0.0
).plot(
    output="pmv",
    thresholds=ThresholdsConfig(thresholds=[-0.5, 0.5]),
    ax=ax0,
    title="Legend Off",
    x_resolution=0.2,
    y_resolution=0.5,
    legend=False,
)

Threshold(model_func=pmv_ppd_iso).x(tdb=(18.0, 34.0)).y(rh=(20.0, 100.0)).parameters(
    vr=0.10, met=1.2, clo=0.5, wme=0.0
).plot(
    output="pmv",
    thresholds=ThresholdsConfig(thresholds=[-0.5, 0.5]),
    ax=ax1,
    title="Boundary Lines Off",
    x_resolution=0.2,
    y_resolution=0.5,
    show_lines=False,
)

plt.show()
```

## UTCI Threshold Example

The same `Threshold` workflow also works with UTCI. In this example, the x-axis is dry-bulb temperature, the y-axis is relative humidity, `v=0.5` is fixed through `.parameters(...)`, and the default-link behavior fills `tr` from `tdb`.

```python
Model_2 = (
    Threshold(model_func=utci)
    .x(tdb=(10.0, 40.0))
    .y(rh=(20.0, 100.0))
    .parameters(v=1.0)
    .plot(
        output="utci",
        thresholds=ThresholdsConfig(
            thresholds=[26, 32],
            labels=["Cool", "Comfortable", "Hot"],
            colors=["#4c78a8", "#d9d9d9", "#c9415f"],
        ),
        title="UTCI Threshold Regions",
        x_resolution=1.0,
        y_resolution=0.5,
        line_kws={"linewidth": 2.0, "linestyle": "--"},
        fill_kws={"alpha": 0.65},
    )
)
Model_2.ax.set_title("UTCI Threshold-Filled Regions")
Model_2.ax.set_xlabel("Air temperature [degC]")
Model_2.ax.set_ylabel("Relative humidity [%]")
```
